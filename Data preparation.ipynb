```python
import numpy as np
import os
import matplotlib.pyplot as plt
import cv2
from tqdm import tqdm
```

```python
TARGETFOLDER = "img_align_celeba"
```

```python
currentFolder = os.listdir(TARGETFOLDER)
```

```python
%matplotlib inline
```

```python
img  = cv2.imread(os.path.join(TARGETFOLDER, currentFolder[0]))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
train = []
```

```python
for i in tqdm(currentFolder):
    img = cv2.imread(os.path.join(TARGETFOLDER, i))
    img  = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (64, 64))
    train.append(img)
```

```python
train = np.array(train)
```

```python
plt.figure(figsize=(20,30))
for i in range(15):
    plt.subplot(5, 3, i + 1)
    plt.imshow(train[i])
```

```python
np.save("Face_Dataset_64RGB", train)
```
